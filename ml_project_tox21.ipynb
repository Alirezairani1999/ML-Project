Oversampling using SMOTE
one of the oversampling technique is SMOTE (Synthetic Minority Over-sampling TEchnique).
It is a more sophisticated technique than the other ones. 
Random sampling is easy, but the new samples don’t add more information to the machine learning algorithms. SMOTE improves on that.
SMOTE oversamples the minority class by creating ‘synthetic’ examples rather than copies. 
It involves some methods, including nearest neighbors, to generate plausible new examples.

Undersampling using Tomek links

This method detects Tomek links and removes samples based on them.
What is the Tomek link?
It is between two samples of different classes. When the two samples are the nearest neighbors of each other, they form a Tomek link.
 we remove the sample from the majority class to achieve undersampling, i.e., remove the majority class close to the minority class. This removes ambiguity between the two classes.

SMOTE and Tomek links
The SMOTE oversampling approach could generate noisy samples since it creates synthetic data. 
To solve this problem, after SMOTE, we could use undersampling techniques to clean up.


```python
from imblearn.combine import SMOTETomek

# summarize class distribution
print("Before rsampling: ",Counter(Y_train))

# define oversampling strategy
smote_tomek = SMOTETomek(random_state=888)

# fit and apply the transform
X_resampled, y_resampled = smote_tomek.fit_resample(X_train_scaled, Y_train)

# summarize class distribution
print("After rsampling: ",Counter(y_resampled))
```

#################### Logistic Regresssion ##########################

```python
# define model
clf = LogisticRegression()
clf.fit(X_resampled, y_resampled)
```

```python

# define pipeline
pipeline = Pipeline(steps=[('r', smote_tomek), ('m', clf)])
# define evaluation procedure
cv = StratifiedKFold(shuffle=True)
# evaluate model
scores_clf = cross_val_score(pipeline, X_train_scaled, Y_train==1, scoring='roc_auc', cv=cv, n_jobs=-1)
# summarize performance
print('Mean ROC AUC: %.3f' % mean(scores_clf))
```

```python
#test set
y_pred_clf = clf.predict_proba(X_test_scaled)[:, 1]
roc_auc_score(Y_test, y_pred_clf)
```

```python
# counfusion Matrix
predict_clf= clf.predict(X_test_scaled)
print(confusion_matrix(Y_test, predict_clf))
print("\n")
print(classification_report(Y_test, predict_clf))
```
